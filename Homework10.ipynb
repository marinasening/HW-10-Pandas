```python
import pandas as pd
```

```python
#Задание 1

movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
movies_rating = movies.merge(ratings, how = 'inner', on = 'movieId')
movies_rating.drop(['genres', 'userId', 'timestamp'], axis = 1, inplace = True)
best_movies = movies_rating[movies_rating.rating == 5]
best_movies_rating = best_movies.groupby('title')['rating'].count().sort_values(ascending = False)
#bmr = best_movies_rating.sort_values('rating', ascending = True).head()
#merged.to_csv('movie_raiting')
best_movies_rating.head(1)
```
